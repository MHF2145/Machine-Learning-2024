### ***Data Cleaning***

Menghapus kolom yang tidak relevan.

```python
df = df.drop(['model', 'trim', 'transmission', 'vin', 'interior', 'saledate'], axis=1)
df.info()
```

Menangani nilai kosong pada kolom numerik.

Pertama, kita melakukan visualisasi data untuk mengevaluasi apakah kita harus memilih rata-rata (mean) atau median sebagai metode penanganan nilai kosong.

```python
plt.figure(figsize=(15, 10))

plt.subplot(4, 1, 1)
plt.hist(df['condition'].dropna(), bins=20, color='blue', alpha=0.5)
plt.title('Distribusi Condition')
plt.xlabel('Condition')

plt.subplot(4, 1, 2)
plt.hist(df['odometer'].dropna(), bins=20, color='green', alpha=0.5)
plt.title('Distribusi Odometer')
plt.xlabel('Odometer')

plt.subplot(4, 1, 3)
plt.hist(df['mmr'].dropna(), bins=20, color='yellow', alpha=0.5)
plt.title('Distribusi MMR')
plt.xlabel('Odometer')

plt.subplot(4, 1, 4)
plt.hist(df['sellingprice'].dropna(), bins=20, color='orange', alpha=0.5)
plt.title('Distribusi Selling Price')
plt.xlabel('Selling Price')

plt.tight_layout()
plt.show()
```

Karena variasi data pada atribut Condition lebih signifikan, maka nilai kosong diisi dengan rata-rata (mean). Sementara itu, untuk atribut Odometer, MMR, dan Selling Price, yang memiliki beberapa pencilan (outlier), nilai kosong akan diisi dengan median.

```python
df['condition'].fillna(df['condition'].mean(), inplace=True)
df['odometer'].fillna(df['odometer'].median(), inplace=True)
df['mmr'].fillna(df['mmr'].median(), inplace=True)
df['sellingprice'].fillna(df['sellingprice'].median(), inplace=True)
```

Menangani nilai null pada kolom kategorikal.

Salah satu cara untuk melakukan ini adalah dengan menggantinya dengan kategori placeholder, seperti 'Other'.

```python
df['make'].fillna('Other', inplace=True)
df['body'].fillna('Other', inplace=True)
df['color'].fillna('Other', inplace=True)
```

***Data Preprocessing***

Dalam tahap ini, diambil sampel sebanyak 10.000 data dari total 500.000 data untuk mempercepat proses komputasi.

```python
from sklearn.preprocessing import StandardScaler

sampled_data = df.sample(n=10000, replace=False)

scaler = StandardScaler()
scaled_df = scaler.fit_transform(sampled_data[numeric_columns])
scaled_df
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=3)

# Fit and transform the data
df_pca = pca.fit_transform(scaled_df)
df_pca
```

### ***Hierarchical Clustering***
##### ***Silhouette Score***

```python
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import AgglomerativeClustering
from scipy.spatial.distance import pdist, squareform
from sklearn.metrics import silhouette_score
from sklearn.metrics import davies_bouldin_score

X = df_pca
cluster_range = range(2, 11)
silhouette_scores = []

for n_clusters in cluster_range:
    hierarchical_clustering = AgglomerativeClustering(n_clusters=n_clusters)
    cluster_labels = hierarchical_clustering.fit_predict(X)
    silhouette_avg = silhouette_score(X, cluster_labels)
    silhouette_scores.append(silhouette_avg)

plt.plot(cluster_range, silhouette_scores)
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score vs. Number of Clusters')
plt.show()
```

Dapat diamati bahwa jumlah klaster terbaik adalah 3.

```python
# Compute the pairwise distances between samples
dist_matrix = squareform(pdist(X))

# Fit the Agglomerative Clustering model
agg_cluster = AgglomerativeClustering(n_clusters=3)
agg_cluster.fit(scaled_df)

# Plot the data points colored by their cluster labels in 2D
plt.scatter(X[:, 0], X[:, 1], c=agg_cluster.labels_, cmap='viridis')
plt.show()

# Plot the data points colored by their cluster labels in 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=agg_cluster.labels_, cmap='viridis')
plt.show()
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt

# Compute the linkage matrix
Z = linkage(X, method='ward')

# Plot the dendrogram
plt.figure(figsize=(12, 6))
dendrogram(Z)
plt.title('Hierarchical Clustering Dendrogram, Ward Link')
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.show()
```

```python
Z = linkage(X, method='single')  

plt.figure(figsize=(15, 8))
dendrogram(Z)
plt.title('Hierarchical Clustering Dendrogram, Single Link')
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.show()
```

```python
Z = linkage(X, method='complete')  

plt.figure(figsize=(12, 6))
dendrogram(Z)
plt.title('Hierarchical Clustering Dendrogram, Complete Link')
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.show()
```

```python
Z = linkage(X, method='average')  

plt.figure(figsize=(12, 6))
dendrogram(Z)
plt.title('Hierarchical Clustering Dendrogram, Average Link')
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.show()
```

Dapat disimpulkan bahwa metode penggabungan (linkage) yang optimal adalah Complete Link.
