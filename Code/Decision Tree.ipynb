```python
#import library
from sklearn.datasets import load_iris #iris dataset dari skleran
from sklearn import tree #library decision tree
import matplotlib.pyplot as plt #library matplotlib
```

```python
#load data
iris=load_iris()
```

```python
iris #dataset iris
```

```python
#mengambil data dan target dari dataset iris
X=iris.data 
y=iris.target
```

```python
#melihat nilai dari variable x
X[:5]
```

```python
#melihat nilai dari variable y/target
y[:5]
```

```python
#kodingan di bawah ini digunakan untuk fit model
# untuk parameter bisa diubah menggunakan gini/entropi serta maxdepth yang diinginkan
clf = tree.DecisionTreeClassifier(criterion='gini',max_depth=5)
# clf = tree.DecisionTreeClassifier(criterion='entropi',max_depth=5)
# clf = tree.DecisionTreeClassifier(criterion='gini',max_depth=2)
clf = clf.fit(X, y)
#Parameter Decision tree lainnya bisa dicek di
##https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeClassifier.html#sklearn.tree.DecisionTreeClassifier
```

```python
plt.figure(figsize=(20, 20))
tree.plot_tree(clf,feature_names=iris.feature_names,  
                      class_names=iris.target_names,  
                      filled=True, rounded=True,) 
```

# GraphViz untuk menyimpan output graph


#### install graphviz terlebih dahulu, bisa menggunakan conda install / pip
#### conda install python-graphviz atau pip install graphviz

```python
import graphviz 
```

```python
# export data ke dalam sebuah file pdf
dot_data = tree.export_graphviz(clf, out_file=None, 
                      feature_names=iris.feature_names,  
                      class_names=iris.target_names,  
                      filled=True, rounded=True,  
                      special_characters=True)
graph = graphviz.Source(dot_data) 
graph.render("iris") 
```
