Kesimpulan dan Analisis Hasil K-Means

Setelah melakukan klasifikasi k-means, analisis dapat dilakukan dengan memeriksa statistik nilai rata-rata per cluster dan mengevaluasi visualisasinya:

Cluster Pertama:

    a. Pendidikan: Tingkat pendidikan cenderung lebih rendah daripada cluster kedua.

    b. Pendapatan: Pendapatan rendah.

    c. Terakhir Berbelanja: Biasanya berbelanja dalam 1-2 bulan terakhir.

    d. Kunjungan Bulanan ke Website: Mengunjungi situs web lebih sering.

    e. Usia: Sedikit lebih muda dibandingkan dengan cluster kedua.

    f. Barang yang Dibeli: Pembelian barang sedikit.

    g. Ukuran Keluarga: Jumlah keluarga relatif lebih banyak.

    h. Total Pembelian: Jarang melakukan transaksi.

Cluster Kedua:

    a. Pendidikan: Tingkat pendidikan cenderung lebih tinggi daripada cluster pertama.

    b. Pendapatan: Pendapatan tinggi.

    c. Terakhir Berbelanja: Biasanya berbelanja dalam 1-2 bulan terakhir.

    d. Kunjungan Bulanan ke Website: Mengunjungi situs web lebih jarang.

    e. Usia: Sedikit lebih tua dibandingkan dengan cluster pertama.

    f. Barang yang Dibeli: Pembelian barang banyak.

    g. Ukuran Keluarga: Jumlah keluarga relatif lebih sedikit.

    h. Total Pembelian: Sering melakukan transaksi.

    Berikut adalah visualisasi fitur antar cluster setelah dibagi menjadi 2 cluster.

```python
df.drop(["ID","Year_Birth","Z_CostContact","Z_Revenue","Dt_Customer"], axis=1, inplace=True)
```

```python
education_ctgrs = pd.Categorical(df["Education"], categories=["Basic", "Graduation", "2n Cycle", "Master", "PhD"], ordered=True)
df["educations"], CAT = pd.factorize(education_ctgrs, sort=True)
```

```python
df[["Education","educations"]]
```

```python
df.drop(["Education"],axis=1, inplace=True)
df.head()
```

```python
df = df.replace(to_replace="Single",value=0)
df = df.replace(to_replace="Divorced",value=0)
df = df.replace(to_replace="Widow",value=0)
df = df.replace(to_replace="Alone",value=0)
df = df.replace(to_replace="Absurd",value=0)
df = df.replace(to_replace="YOLO",value=0)
df = df.replace(to_replace="Married",value=1)
df = df.replace(to_replace="Together",value=1)
```

```python
df.isna().sum()
```

```python
df['Income'].fillna(df['Income'].mean(), inplace=True)
```

```python
df = df.drop(df[df.Income > 400000].i)
```

```python
pca = PCA(n_components=2, whiten=True)
pca.fit(df)
data_pca = pca.transform(df)
print("Explained Variance Ratios: ", pca.explained_variance_ratio_)
print("Total Explained Variance: ", sum(pca.explained_variance_ratio_))
```

```python
data_pca
```

```python
wcss = []
for k in range(1, 15):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(data_pca)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 15), wcss)
plt.xlabel("Number of Clusters (k)")
plt.xticks(range(1, 15, 1))
plt.ylabel("Within-Cluster-Sum-of-Squares (WCSS) Value")
plt.title("Elbow Method for Optimal k Selection")
plt.show()
```

```python
kmeans2 = KMeans(n_clusters=3)
clusters = kmeans2.fit_predict(data_pca)
```

```python
colors = plt.cm.get_cmap('tab10', 3)
for cluster_num in range(3):
    plt.scatter(data_pca[clusters == cluster_num, 0], 
                data_pca[clusters == cluster_num, 1], 
                label=f'{cluster_num + 1}. Customer Cluster', 
                color=colors(cluster_num))

plt.title('Customer Clusters')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.legend()
plt.show()
```
